```python
# Implement a code to find the second largest nubber in a given list of integers

def find_second_largest(numbers):
    if len(numbers) < 2:
        raise ValueError("List must contain at least two distinct numbers.")
    
    # Initialize the largest and second largest
    largest = second_largest = float('-inf')
    
    for number in numbers:
        if number > largest:
            second_largest = largest
            largest = number
        elif largest > number > second_largest:
            second_largest = number
    
    if second_largest == float('-inf'):
        raise ValueError("No distinct second largest number found.")
    
    return second_largest

# Example 
numbers = [10, 20, 4, 45, 99, 99, 20]
print("Second largest number is:", find_second_largest(numbers))  # Output: 45
```

```python
# Create a code to count the occurrences of each element in a list and return a dictionary with element as key and their count as values.

def count_occurrences(lst):
    # Initialize an empty dictionary to store the counts
    counts = {}
    
    # Iterate through each element in the list
    for element in lst:
        # Increment the count for the element or initialize it if not present
        if element in counts:
            counts[element] += 1
        else:
            counts[element] = 1
    
    return counts

# Example 
example_list = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
result = count_occurrences(example_list)
print(result)  # Output: {1: 1, 2: 2, 3: 3, 4: 4}
```

```python
# Write a code to reverse a list in-place without using any built-in reverse functions.

def reverse_list_in_place(lst):
    left = 0
    right = len(lst) - 1
    
    while left < right:
        # Swap elements at the left and right pointers
        lst[left], lst[right] = lst[right], lst[left]
        
        # Move the pointers towards the center
        left += 1
        right -= 1

# Example usage
example_list = [1, 2, 3, 4, 5]
reverse_list_in_place(example_list)
print(example_list)  # Output: [5, 4, 3, 2, 1]
```

```python
# Implement a code to find and remove duplicates from a list while preserving the original order of elements.

def remove_duplicates(lst):
    seen = set()
    unique_list = []
    
    for item in lst:
        if item not in seen:
            seen.add(item)
            unique_list.append(item)
    
    return unique_list

# Example usage
example_list = [1, 2, 3, 2, 4, 1, 5, 6, 4]
result = remove_duplicates(example_list)
print(result)  # Output: [1, 2, 3, 4, 5, 6]
```

```python
# Create a code to check if a given list is sorted (either in 'ascending or desdending order) or not.

def is_sorted(lst):
    # Check if the list is sorted in ascending order
    def is_ascending(lst):
        return all(lst[i] <= lst[i + 1] for i in range(len(lst) - 1))
    
    # Check if the list is sorted in descending order
    def is_descending(lst):
        return all(lst[i] >= lst[i + 1] for i in range(len(lst) - 1))
    
    # Edge case: Empty list or single-element list
    if len(lst) < 2:
        return True
    
    # Determine if the list is sorted in ascending or descending order
    if is_ascending(lst):
        return 'ascending'
    elif is_descending(lst):
        return 'descending'
    else:
        return 'not sorted'

# Example usage
list1 = [1, 2, 3, 4, 5]
list2 = [5, 4, 3, 2, 1]
list3 = [1, 3, 2, 4, 5]
list4 = []

print(f"List {list1} is {is_sorted(list1)}")  # Output: ascending
print(f"List {list2} is {is_sorted(list2)}")  # Output: descending
print(f"List {list3} is {is_sorted(list3)}")  # Output: not sorted
print(f"List {list4} is {is_sorted(list4)}")  # Output: True
```

```python
# Write a code to merge two sorted lists into a single sorted list.

def merge_sorted_lists(list1, list2):
    # Initialize pointers for both lists and the merged list
    i, j = 0, 0
    merged_list = []
    
    # Traverse both lists and append the smaller element to the merged list
    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1
    
    # Append any remaining elements from list1
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1
    
    # Append any remaining elements from list2
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1
    
    return merged_list

# Example usage
list1 = [1, 3, 5, 7]
list2 = [2, 4, 6, 8]
result = merge_sorted_lists(list1, list2)
print(result)  # Output: [1, 2, 3, 4, 5, 6, 7, 8]
```

```python
# Implement a code to find the intersection of two given lists.

def find_intersection(list1, list2):
    # Convert lists to sets to leverage set operations
    set1 = set(list1)
    set2 = set(list2)
    
    # Find the intersection of both sets
    intersection = set1 & set2
    
    # Convert the result back to a list if needed
    return list(intersection)

# Example usage
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
result = find_intersection(list1, list2)
print(result)  # Output: [4, 5]
```

```python
# Write a code to shuffle a given list randomly without using any built-in shuffle functions.

import random

def fisher_yates_shuffle(lst):
    # Traverse the list in reverse order
    for i in range(len(lst) - 1, 0, -1):
        # Generate a random index from 0 to i (inclusive)
        j = random.randint(0, i)
        
        # Swap elements at indices i and j
        lst[i], lst[j] = lst[j], lst[i]

# Example usage
example_list = [1, 2, 3, 4, 5]
fisher_yates_shuffle(example_list)
print(example_list)  # Output: The list in a randomly shuffled order
```

```python
# Write a code that takes a list of words as input and returns a diction+ry where the keys are unique words and the values are the frequencies of those words in the input list.

def count_word_frequencies(words_list):
    # Initialize an empty dictionary to store word frequencies
    frequency_dict = {}
    
    # Iterate over each word in the list
    for word in words_list:
        # Convert word to lowercase to ensure case insensitivity
        word = word.lower()
        # Update the frequency count for the word
        if word in frequency_dict:
            frequency_dict[word] += 1
        else:
            frequency_dict[word] = 1
    
    return frequency_dict

# Example usage
input_list = ["apple", "banana", "apple", "orange", "banana", "banana"]

# Get the word frequencies
word_frequencies = count_word_frequencies(input_list)

# Print the result
print("Word frequencies:", word_frequencies)
```

```python
# Write a code that takes two dictionaries as input and merges them into a single dictionary. If there are common keys the values should be added together.

def merge_dicts(dict1, dict2):
    # Create a new dictionary to store the merged result
    merged_dict = {}
    
    # Iterate over the first dictionary and add its items to the merged dictionary
    for key in dict1:
        if key in merged_dict:
            merged_dict[key] += dict1[key]
        else:
            merged_dict[key] = dict1[key]
    
    # Iterate over the second dictionary and add its items to the merged dictionary
    for key in dict2:
        if key in merged_dict:
            merged_dict[key] += dict2[key]
        else:
            merged_dict[key] = dict2[key]
    
    return merged_dict

# Example usage
dict1 = {'a': 10, 'b': 20, 'c': 30}
dict2 = {'b': 15, 'c': 25, 'd': 5}

# Merge the dictionaries
result_dict = merge_dicts(dict1, dict2)

# Print the result
print("Merged dictionary:", result_dict)
```

```python
# Write a codee to access a value in a nested dictionary. The function should take the dictionary and a list of keys as input and return the corresponding value. If any of the keys do not exist in the dictionary, the function should return None.

def get_nested_value(d, keys):
    # Initialize the current level of the dictionary
    current_level = d
    
    # Iterate through the list of keys
    for key in keys:
        # Check if the current key exists in the current level of the dictionary
        if key in current_level:
            # Move to the next level in the nested dictionary
            current_level = current_level[key]
        else:
            # Return None if the key does not exist
            return None
    
    # Return the final value if all keys are found
    return current_level

# Example usage
nested_dict = {
    'a': {
        'b': {
            'c': 42
        }
    },
    'x': {
        'y': {
            'z': 99
        }
    }
}

keys_list1 = ['a', 'b', 'c']
keys_list2 = ['x', 'y', 'z']
keys_list3 = ['a', 'b', 'd']  # This key does not exist

# Access values using the function
value1 = get_nested_value(nested_dict, keys_list1)
value2 = get_nested_value(nested_dict, keys_list2)
value3 = get_nested_value(nested_dict, keys_list3)

# Print the results
print("Value for keys_list1:", value1)  # Should print: 42
print("Value for keys_list2:", value2)  # Should print: 99
print("Value for keys_list3:", value3)  # Should print: None
```

```python
# Write a code that takes a dictionary as input and returns a sorted version of it based on the values. You can choose whether to sort in ascending or descending order.

def sort_dict_by_values(d, ascending=True):
    """
    Sort a dictionary by its values.
    
    :param d: Dictionary to be sorted.
    :param ascending: Boolean indicating sorting order; True for ascending, False for descending.
    :return: A dictionary sorted by its values.
    """
    # Sort dictionary items based on values
    sorted_items = sorted(d.items(), key=lambda item: item[1], reverse=not ascending)
    
    # Convert the sorted items back into a dictionary
    sorted_dict = dict(sorted_items)
    
    return sorted_dict

# Example usage
example_dict = {'apple': 10, 'banana': 5, 'cherry': 7, 'date': 3}

# Sort in ascending order
sorted_ascending = sort_dict_by_values(example_dict, ascending=True)
print("Sorted in ascending order:", sorted_ascending)

# Sort in descending order
sorted_descending = sort_dict_by_values(example_dict, ascending=False)
print("Sorted in descending order:", sorted_descending)

```
